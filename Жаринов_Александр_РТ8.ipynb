–¢—Ä–µ–±—É–µ—Ç—Å—è –ø—Ä–æ–≤–µ—Å—Ç–∏ –∫–ª–∞—Å—Ç–µ—Ä–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–º k-—Å—Ä–µ–¥–Ω–∏—Ö.
–ü–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([[5,3],
              [10,15],
              [15,12],
              [24,10],
              [30,45],
              [85,70],
              [71,80],
              [60,78],
              [55,52],
              [80,91]])

# –ø—Ä–æ–±—É–µ–º —Å —Ç—Ä–µ–º—è –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# –ø–µ—á–∞—Ç–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏
print("–î–ª—è —Ç—Ä—ë—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤")
print("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:")
print(kmeans.predict(X))

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã
plt.figure()
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()

# –ø—Ä–æ–±—É–µ–º —Å –¥–≤—É–º—è –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# –ø–µ—á–∞—Ç–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏
print("–î–ª—è –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤")
print("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:")
print(kmeans.predict(X))

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã
plt.figure()
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()

# –ø—Ä–æ–±—É–µ–º —Å —á–µ—Ç—ã—Ä—å–º—è –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# –ø–µ—á–∞—Ç–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏
print("–î–ª—è —á–µ—Ç—ã—Ä—ë—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤")
print("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:")
print(kmeans.predict(X))

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã
plt.figure()
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()
```

–£—Å–ª–æ–≤–∏–µ 1.2: –í—ã–ø–æ–ª–Ω–∏—Ç–µ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é –¥–ª—è –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –∏—Ä–∏—Å–æ–≤ –§–∏—à–µ—Ä–∞. –í—ã–ø–æ–ª–Ω–∏—Ç–µ
–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏. –ü–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# –∑–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
iris = datasets.load_iris()
X = iris.data

# –ø—Ä–æ–±—É–µ–º —Å —Ç—Ä–µ–º—è –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# –ø–µ—á–∞—Ç–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏
print("–î–ª—è —Ç—Ä—ë—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤")
print("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:")
print(kmeans.predict(X))

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã
plt.figure()
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()

# –ø—Ä–æ–±—É–µ–º —Å –¥–≤—É–º—è –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏
print("–î–ª—è –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤")
print("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:")
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# –ø–µ—á–∞—Ç–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏
print(kmeans.predict(X))

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã
plt.figure()
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()

# –ø—Ä–æ–±—É–µ–º —Å —á–µ—Ç—ã—Ä—å–º—è –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# –ø–µ—á–∞—Ç–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏
print("–î–ª—è —á–µ—Ç—ã—Ä—ë—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤")
print("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:")
print(kmeans.predict(X))

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã
plt.figure()
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()

# –Ω–∞—Ö–æ–¥–∏–º –∫–ª–∞—Å—Ç–µ—Ä—ã –¥–ª—è –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–∏ –¥–∞–Ω–Ω—ã—Ö
clusters = pairwise_distances_argmin_min(X, kmeans.cluster_centers_)
print(clusters)
```

**–ó–∞–¥–∞–Ω–∏–µ 2: –ò–µ—Ä–∞—Ä—Ö–∏—á–µ—Å–∫–∞—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è**

–£—Å–ª–æ–≤–∏–µ: –ó–∞–¥–∞–Ω–∏–µ. –í—ã–ø–æ–ª–Ω–∏—Ç–µ –∏–µ—Ä–∞—Ä—Ö–∏—á–µ—Å–∫—É—é –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é –¥–ª—è –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –æ–±
–∏—Ä–∏—Å–∞—Ö –§–∏—à–µ—Ä–∞. –ü—Ä–∏ —ç—Ç–æ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ª—é–±—ã–µ –¥–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞
(–≤—Å–µ–≥–æ –∏—Ö —á–µ—Ç—ã—Ä–µ).

```python
#–ü—Ä–∏–∑–Ω–∞–∫–∏ - –¥–ª–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞ –∏ –¥–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])

X = df[['sepal length (cm)', 'petal length (cm)']]

Z = linkage(X, 'ward')

plt.figure(figsize=(10, 7))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
dendrogram(Z)
plt.show()
```
