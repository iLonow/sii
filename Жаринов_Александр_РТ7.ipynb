```python
import numpy as np

class OurNeuralNetwork1:
    def __init__(self):
        self.weights = np.array([0.5, 0.5, 0.5])
        self.bias = 0

    def feedforward(self, x1, x2, x3):
        sum = np.dot(self.weights, np.array([x1, x2, x3])) + self.bias
        return self.sigmoid(sum)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

class OurNeuralNetwork2:
    def __init__(self):
        self.weights = np.array([1, 0])
        self.bias = 1

    def feedforward(self, x1, x2):
        sum = np.dot(self.weights, np.array([x1, x2])) + self.bias
        return self.sigmoid(sum)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

network1 = OurNeuralNetwork1()
network2 = OurNeuralNetwork2()
print("–ü–µ—Ä–≤–∞—è –Ω–µ–π—Ä–æ—Å–µ—Ç—å")
print(network1.feedforward(0, 10, 100))
print("–í—Ç–æ—Ä–∞—è –Ω–µ–π—Ä–æ—Å–µ—Ç—å")
print(network2.feedforward(0, 10))
print(network2.feedforward(0, 20))
```

```python
import numpy as np

class OurNeuralNetwork:
    def __init__(self):
        self.weights = np.array([0.5, 0.5, 0.5])
        self.bias = 0

    def feedforward(self, x1, x2, x3):
        sum = np.dot(self.weights, np.array([x1, x2, x3])) + self.bias
        return self.tanh(sum)

    def tanh(self, x):
        return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

class OurNeuralNetwork2:
    def __init__(self):
        self.weights = np.array([1, 0])
        self.bias = 1

    def feedforward(self, x1, x2):
        sum = np.dot(self.weights, np.array([x1, x2])) + self.bias
        return self.tanh(sum)

    def tanh(self, x):
        return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

network = OurNeuralNetwork()
network2 = OurNeuralNetwork2()
print("–ü–µ—Ä–≤–∞—è –Ω–µ–π—Ä–æ—Å–µ—Ç—å")
print(network.feedforward(0, 10, 100))
print("–í—Ç–æ—Ä–∞—è –Ω–µ–π—Ä–æ—Å–µ—Ç—å")
print(network2.feedforward(0, 10))
print(network2.feedforward(0, 20))
```

```python
import numpy as np

class OurNeuralNetwork1:
    def __init__(self):
        self.weights = np.array([0.5, 0.5, 0.5])
        self.bias = 0

    def feedforward(self, x1, x2, x3):
        sum = np.dot(self.weights, np.array([x1, x2, x3])) + self.bias
        return self.relu(sum)

    def relu(self, x):
        return np.maximum(0, x)

class OurNeuralNetwork2:
    def __init__(self):
        self.weights = np.array([1, 0])
        self.bias = 1

    def feedforward(self, x1, x2):
        sum = np.dot(self.weights, np.array([x1, x2])) + self.bias
        return self.relu(sum)

    def relu(self, x):
        return np.maximum(0, x)

network1 = OurNeuralNetwork1()
network2 = OurNeuralNetwork2()
print("–ü–µ—Ä–≤–∞—è –Ω–µ–π—Ä–æ—Å–µ—Ç—å")
print(network1.feedforward(0, 10, 100))
print("–í—Ç–æ—Ä–∞—è –Ω–µ–π—Ä–æ—Å–µ—Ç—å")
print(network2.feedforward(0, 10))
print(network2.feedforward(0, 20))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix

url = r'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
dataset = pd.read_csv(url)

X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.80, test_size = 0.20, stratify = Y, random_state = 123)

mlp_classifier = MLPClassifier(max_iter = 1000).fit(X_train, Y_train)
Y_pred = mlp_classifier.predict(X_test)

df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
print(df)

print("\nTest Accuracy: %.3f"%mlp_classifier.score(X_test, Y_test))
print('Training Accuracy: %.3f'%mlp_classifier.score(X_train, Y_train))
print(f"\nMatrix of Confusion:\n {confusion_matrix(Y_test, Y_pred)}")
print(f"\nLoss : ", mlp_classifier.loss_)
print(f'Number of Coefs : ', len(mlp_classifier.coefs_))
print(f'Number of Intercepts : ', len(mlp_classifier.intercepts_))
print(f'Number of Iterations for Which Estimator Run : ', mlp_classifier.n_iter_)
print(f'Name of Output Layer Activation Function : ', mlp_classifier.out_activation_)
```
