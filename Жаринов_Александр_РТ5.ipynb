Задание 2
Представьте дерево показанное на рисунке с использованием списка из
списков. Выведите на печать корень дерева, а также его левое и правое
поддеревья

```python
class Tree:
    def __init__(self, kids, next=None):
        self.kids = self.val = kids
        self.next = next
t = Tree(Tree('a', Tree('b', Tree('d', 'e'))), Tree('c', Tree('f')))
print('Root = ', t.kids.val)
print('Left subtree = ', end='')
print(t.kids.next.val, end=', ')
print(t.kids.next.next.val, end=', ')
print(t.kids.next.next.next)
print('Right subtree = ', end='')
print(t.next.kids, end=', ')
print(t.next.next.kids)
```

Реализуйте в классе функцию для вставки нового элемента в дерево по
следующим правилам:
• Левое поддерево узла содержит только узлы со значениями меньше,
чем значение в узле.
• Правое поддерево узла содержит только узлы со значениями меньше,
чем значение в узле.
• Каждое из левого и правого поддеревьев также должно быть
бинарным деревом поиска.
• Не должно быть повторяющихся узлов.
Метод вставки сравнивает значение узла с родительским узлом и решает
куда доваить элемент (в левое или правое поддерево). Перепишите, метод
PrintTree для печати полной версии дерева

```python
class Tree:
    def __init__(self, d):
        self.left = None
        self.right = None
        self.d = d

    def i(self, d):
        if self.d is None:
            self.d = d
        else:
            if d < self.d:
                if self.left is None:
                    self.left = Tree(d)
                else:
                    self.left.i(d)
            else:
                if self.right is None:
                    self.right = Tree(d)
                else:
                    self.right.i(d)

    def PrintTree(self, prefix="", is_left=True):
        if self.left:
            self.left.PrintTree(prefix + "     ", False)
        print(prefix + ("   " if is_left else "   ") + str(self.d))
        if self.right:
            self.right.PrintTree(prefix + "     ", True)
tree = Tree(50)
tree.i(30)
tree.i(20)
tree.i(40)
tree.i(70)
tree.i(60)
tree.i(80)
tree.PrintTree()
```

Задание 4
Постройте классификатор на основе дерева принятия решений
следующего датасета:

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

X = np.array([[-1,-1],[-2,-1],[-3,-2],[1,1],[2,1],[3,2]])
target=[0,0,0,1,1,1]

clf = DecisionTreeClassifier()
clf = clf.fit(X, target)
print(clf.predict([[-0.5, 0.5], [2.5, 2.5]]))

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(3, 3), dpi=300)
plot_tree(clf, filled=True)
plt.show()
```

Постройте модель регрессии для данных из предыдущей рабочей

Постройте прогноз. Оцените точность модели.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

url = r'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
dataset = pd.read_csv(url)
print(dataset.head())
print(dataset.shape)
dataset.describe()
plt.scatter(dataset['Petrol_tax'], dataset['Average_income'], color='b', label='Средний доход')
plt.xlabel('Средний доход')
plt.ylabel('Налог на бензин')
plt.title('Прогноз потребления газа')
plt.show()

from sklearn.tree import DecisionTreeRegressor
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(x)
print(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)

from sklearn import tree
tree.plot_tree(regressor)
y_pred = regressor.predict(x_test)
y_pred
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df

print('Средняя абсолютная погрешность:', metrics.mean_absolute_error(y_test, y_pred))
print('Средняя квадратичная погрешность:', metrics.mean_squared_error(y_test, y_pred))
print('Корень средней квадратичной погрешности:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
